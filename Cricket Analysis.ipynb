```python
# Import the library
import pandas as pd
```

```python
df=pd.read_csv('cricket_match.csv.csv', header=1, delimiter=';')
```

```python
#rename the columns for convenient names
df=df.rename(columns={'Mat':'Matches','NO':'Not_Outs', 'HS':'Highest_Inns_Score', 'BF':'Balls_Faced', 'SR':'Batting_Strike_Rate'})
```

```python
df.isnull().any()
```

```python
# Checks the rows that are Nan
df[df['Balls_Faced'].isna()==1]
```

```python
df['Balls_faced']=df['Balls_Faced'].fillna(0,inplace=True)
```

```python
pd.set_option('display.max_rows',None)
```

```python
# Divide the Span into Rookie_Year and Final_Year and delete the Span
df['Span'].str.split(pat='-')
df['Rookie_Year']=df['Span'].str.split(pat='-').str[0]
df['Final_Year']=df['Span'].str.split(pat='-').str[1]
df.drop('Span',axis=1)
```

```python
df=df.drop(columns=['Balls_faced','Span'])
```

```python
df
```

```python
#Divide the Player column into index, name and Country, then delete the first two
df[['index', 'Name','Country']]=df['Player'].str.split(' ',2, expand=True)
df['Country']=df['Country'].str.split(pat='(').str[1]
df['Country']=df['Country'].str.split(pat=')').str[0]
df=df.drop(columns=['Name','index'])
```

```python
#Clean the data by deleting unnecessary parts
df['Highest_Inns_Score']=df['Highest_Inns_Score'].str.split('*').str[0]
df['Balls_Faced']=df['Balls_Faced'].str.split('+').str[0]
df['4s']=df['4s'].str.split('+').str[0]

df
```

```python
df['Balls_faced']=df['Balls_Faced'].fillna(0,inplace=True)
```

```python
df
```

```python
df=df.drop(columns='Balls_faced')
```

```python
# Remove non-numeric characters from '4s' column
df['4s'] = df['4s'].str.replace(r'[^0-9]', '', regex=True)

# Convert to integer
df['4s'] = df['4s'].astype(int)
```

```python
df['6s'] = df['6s'].str.replace(r'[^0-9]', '', regex=True)

# Convert to integer
df['6s'] = df['6s'].astype(int)
```

```python
df.astype
```

```python
# Remove non-numeric characters from 'Balls_Faced' and 'Highest_Inns_Score'
df['Balls_Faced'] = df['Balls_Faced'].str.replace(r'[^0-9]', '', regex=True)
df['Highest_Inns_Score'] = df['Highest_Inns_Score'].str.replace(r'[^0-9]', '', regex=True)
df
```

```python
# Fill NaN values with 0
df['Balls_Faced'] = df['Balls_Faced'].fillna(0)
df['Highest_Inns_Score'] = df['Highest_Inns_Score'].fillna(0)
```

```python
# Convert columns to integers
df['Balls_Faced'] = df['Balls_Faced'].astype(int)
df['Highest_Inns_Score'] = df['Highest_Inns_Score'].astype(int)
```

```python
df.astype({'Rookie_Year':'int','Final_Year':'int'})
```

```python
# Remove non-numeric characters from 'Balls_Faced' and 'Highest_Inns_Score'
df['Rookie_Year'] = df['Rookie_Year'].str.replace(r'[^0-9]', '', regex=True)
df['Final_Year'] = df['Final_Year'].str.replace(r'[^0-9]', '', regex=True)
# Fill NaN values with 0
df['Rookie_Year'] = df['Rookie_Year'].fillna(0)
df['Final_Year'] = df['Final_Year'].fillna(0)

df['Rookie_Year'] = df['Rookie_Year'].astype(int)
df['Final_Year'] = df['Final_Year'].astype(int)
df.dtypes
```

```python
df['Career_Length']=df['Final_Year']-df['Rookie_Year']
```

```python
df['Career_Length'].mean()
```

```python
# average battling strike for players who played over 10 years
df[df['Career_Length']>10]['Batting_Strike_Rate'].mean()
```

```python
# Find number of cricketers who played before 1960
df[df['Final_Year']<1960]['Player'].count()
```

```python
# Max highest Inns score by country
df.groupby('Country')['Highest_Inns_Score'].max().to_frame('Highest_in_country').reset_index().sort_values('Highest_in_country', ascending= False)
```

```python
# 5 Hundreds, fifties, ducks (0) AVG by groupby
df.groupby('Country')[['100','50','0']].mean()
```
